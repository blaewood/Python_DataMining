```python
from sklearn import datasets #load_breast_cancer을 사용하기 위한 sklearn.datasets 라이브러리 호출
from sklearn.model_selection import train_test_split #교차 검증을 위한 라이브러리
from sklearn.tree import DecisionTreeClassifier #결정 트리 
from sklearn.tree import export_graphviz
import pydot
import graphviz

#import numpy  as np
#import pandas as pd
#import matplotlib.pyplot as plt
#from sklearn.model_selection import train_test_split
```

```python
BC=datasets.load_breast_cancer() #BC 객체에 암환자 정보를 전부 받아옴
X, y=datasets.load_breast_cancer(return_X_y=True) #X와 y를 랜덤하게 돌려 분할
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=1, stratify=y)


print(BC.DESCR)
```

```python
# max_depth=10일때의 score 측정
model=DecisionTreeClassifier(max_depth=10, random_state=1).fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
# max_depth=9일때의 score 측정
model=DecisionTreeClassifier(max_depth=9, random_state=1).fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```
