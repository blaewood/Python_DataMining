```python
import pandas as pd
import numpy as np

from sklearn import datasets #Boston Housing을 사용하기 위한 sklearn.datasets 라이브러리 호출
from sklearn.model_selection import train_test_split #교차 검증을 위한 라이브러리
from sklearn.tree import DecisionTreeRegressor #결정 트리 회귀를 사용하기 위한 DecisionTreeRegressor 라이브러리
from sklearn.tree import export_graphviz #결정 트리의 이미지를 추출하기 위한 export_graphviz 라이브러리
import pydot
import graphviz
```

```python
BH=datasets.load_boston()
X, y=datasets.load_boston(return_X_y=True)
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=1) 
#이전 문제와는 다르게stratify 파라미터를 제거해 줘야함, 왜냐하면 무작위 샘플링이어도 train하나에 test도 적어도 하나 이상 존재해야 하는데, 이 데이터에는 특정 라벨 값을 가진 데이터가 없는 경우이다.
print(X.shape)
print(BH.DESCR)
```

```python
# max_depth=10일때의 score 측정
model=DecisionTreeRegressor(max_depth=10, random_state=1).fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```
